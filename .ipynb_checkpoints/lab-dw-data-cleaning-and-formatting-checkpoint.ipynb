## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

```python
import pandas as pd
import numpy as np

csv_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(csv_url)

df.head()
```

```python
df.columns
```

```python
# Standardize the column names
df.columns = df.columns.str.lower()  # Convert all column names to lowercase
df.columns = df.columns.str.replace(" ", "_")  # Replace spaces with underscores

# Display the first few rows of the DataFrame to verify the changes
display(df)
```

```python
df.columns
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they contain only valid data. For example, the gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

```python
df["gender"].unique()
```

```python
# Mapping dictionary to standardize gender values
gender_mapping = {
    "F": "Female",
    "M": "Male",
    "Female": "Female",
    "Male": "Male",
    "female": "Female",
}

# Clean the "gender" column using the mapping dictionary
df["gender"] = df["gender"].map(gender_mapping)
```

```python
df["gender"].unique()
```

```python
#other form to rename

df.replace(to_replace = ["Cali"],
                          value = "California", inplace=True)

df.replace(to_replace = ["Cali"],
                          value = "California", inplace=True)
                          
df.replace(to_replace = ["WA"],
                          value = "Washington ", inplace=True)
                          
df.replace(to_replace = ["AZ"],
                          value = "Arizona", inplace=True)
                          
df.replace(to_replace = ["Bachelors", "College"],
                          value = "Bachelor", inplace=True)
```

```python
df.head()
```
