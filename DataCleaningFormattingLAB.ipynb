# Challenge 2: creating functions on a separate `py` file

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting. 

Write these functions in a separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain. 

*Hint: autoreload module is a utility module in Python that allows you to automatically reload modules in the current session when changes are made to the source code. This can be useful in situations where you are actively developing code and want to see the effects of changes you make without having to constantly restart the Python interpreter or Jupyter Notebook kernel.*

```python
def identify_columns_with_null(data):
    # Identify columns with null values
    columns_with_null = data.columns[data.isnull().any()]
    return columns_with_null

def handle_null_values(data, column, strategy='mean'):
    if strategy == 'mean':
        data[column] = data[column].fillna(data[column].mean())
    elif strategy == 'mode':
        data[column] = data[column].fillna(data[column].mode()[0])
    elif strategy == 'previous':
        data[column] = data[column].fillna(method='ffill')
    elif strategy == 'next':
        data[column] = data[column].fillna(method='bfill')
    return data

def convert_numeric_to_integer(data):
    # Convert numeric variables to integers
    data = data.applymap(lambda x: int(x) if pd.notnull(x) and isinstance(x, float) else x)
    return data

def clean_and_format_data(data):
    # Identify columns with null values
    columns_with_null = identify_columns_with_null(data)

    # Handle null values for each column
    for column in columns_with_null:
        if data[column].dtype == 'object':
            data = handle_null_values(data, column, strategy='mode')
        elif data[column].dtype in ['int', 'float']:
            data = handle_null_values(data, column, strategy='mean')

    # Convert numeric variables to integers
    data = convert_numeric_to_integer(data)

    return data

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'

def main():
    # Read the CSV file into a dataframe
    data = pd.read_csv(url)

    # Clean and format the data
    data_cleaned = clean_and_format_data(data)

    # Save the cleaned dataset to a new CSV file
    data_cleaned.to_csv('cleaned_data.csv', index=False)

if __name__ == '__main__':
    main()
```
